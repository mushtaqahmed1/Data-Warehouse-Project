```python
#Customer table
# 4 missing customer IDs - float
# No missing customer keys - object
# 8 missing first names - object
# 7 missing last names - object
# 7 missing marital status values - object
# 4578 missing gender values - object
# 4 missing creation dates - object
```

```python
df_cust['cst_id'] = df_cust['cst_id'].fillna(00000.0)
```

```python
cols = ['cst_firstname','cst_lastname','cst_marital_status','cst_gndr']
df_cust[cols] = df_cust[cols].fillna('Unknown')
```

```python
df_cust['cst_create_date'] = df_cust['cst_create_date'].fillna('5999-01-01')
```

```python
df_cust.isna().sum()
```

```python
# 2 missing product costs
# 17 missing product line values
# No missing product start dates
# 197 missing product end dates (significant data quality issue)
```
